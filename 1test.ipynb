```python
def calcPE(price, earn):
    pe = price/ earn
    return pe


# หุ้น EA ราคา 52.50 บาท, กำไรต่อหุ้น 1.02 บาท
price = 52.50
earn = 1.02
pe = calcPE(price,earn)
print(f"หุ้นนี้ประมาณ {pe} ปี จะคืนทุนได้")
```

ลองพิจารณา ตัวเลขสองชุดนี้

X = -1, 0, 1, 2, 3, 4

Y = -3, -1, 1, 3, 5, 7

คิดหาความสัมพันธ์ระหว่าง X, Y

จากที่เราคิดนะ X นี่ เพิ่มที่ละ 1 ส่วน Y เพิ่มทีละ 2 
มันก็น่าจะเป็น Y = 2X แต่ เมื่อ X เป็น 0 , Y เป็น -1 

มันก็ควรจะเป็น **Y = 2X - 1**

นี่มันคล้ายวิธีการเรียนของเครื่อง เราจะมาลองเขียนโค้ด TensorFlow โดยให้ neural network คิดหาความสัมพันธ์ให้


```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([Dense(units=1, input_shape=[1])])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
model.fit(xs, ys, epochs=500)
```

```python
print(model.predict([10.0]))
# จริงมันควรจะเท่ากับ Y = 2X - 1 -> 2*10 - 1 = 19
```

ลองดูว่า Network มันเรียนรู้อย่างไร


```python
#สร้าง network แบบใหม่ แต่เหมือนเดิม

l0 = Dense(units=1, input_shape=[1])
model = Sequential([l0])
```

```python
model.compile(optimizer='sgd' , loss='mean_squared_error')
```

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
model.fit(xs, ys, epochs=500)
```

```python
print(model.predict([10.0]))
print("Here is what I learned: {}".format(l0.get_weights()))
```
